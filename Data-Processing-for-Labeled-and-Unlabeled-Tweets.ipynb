*Data Set Description*

This is the sentiment140 dataset. It contains 1,600,000 tweets extracted using the twitter api . The tweets have been annotated (0 = negative, 2 = neutral, 4 = positive) and they can be used to detect sentiment . It contains the following 6 fields:

target: the polarity of the tweet (0 = negative, 2 = neutral, 4 = positive)
ids: The id of the tweet ( 2087)
date: the date of the tweet (Sat May 16 23:58:44 UTC 2009)
flag: The query (lyx). If there is no query, then this value is NO_QUERY.
user: the user that tweeted (robotickilldozr)
text: the text of the tweet (Lyx is cool)
The official link regarding the dataset with resources about how it was generated is here The official paper detailing the approach is here

According to the creators of the dataset:

"Our approach was unique because our training data was automatically created, as opposed to having humans manual annotate tweets. In our approach, we assume that any tweet with positive emoticons, like :), were positive, and tweets with negative emoticons, like :(, were negative. We used the Twitter Search API to collect these tweets by using keyword search"

citation: Go, A., Bhayani, R. and Huang, L., 2009. Twitter sentiment classification using distant supervision. CS224N Project Report, Stanford, 1(2009), p.12

Twitter Flu Annotations
Here you can download the annotated data we used for training influenza classifiers described in the paper below. This is only the annotated training data (about 10K tweets), and not the full set of Twitter data.

Also note that this dataset does not contain the content of the tweets. The tweet IDs are included, which you can use to download the tweet content using the Twitter API.

References
Alex Lamb, Michael J. Paul, Mark Dredze. Separating Fact from Fear: Tracking Flu Infections on Twitter. To appear at the 2013 Conference of the North American Chapter of the Association for Computational Linguistics: Human Language Technologies (NAACL-HLT 2013), Atlanta. June 2013.

```python
tweetLabelFiles_AvI = ["AwarenessVsInfection2009TweetIDs.txt", "AwarenessVsInfection2012TweetIDs.txt"]

tweetLabelFiles_RvN = ["RelatedVsNotRelated2009TweetIDs.txt","RelatedVsNotRelated2012TweetIDs.txt"]

tweetLabelFiles_SvO = ["SelfVsOthers2009TweetIDs.txt", "SelfVsOthers2012TweetIDs.txt"]
```

```python
i=0
aggregateColoradoLabelData_AvI = None
for filename in tweetLabelFiles_AvI:
    coloradoData = pandas.read_csv(filename, sep="\t", header=None)
    coloradoData.columns = ["Tweet_ID","Awareness_Label"]
    #coloradoData.set_index("Tweet_ID")    ////This doesnt work for some reason...
    if i == 0:
        aggregateColoradoLabelData_AvI = coloradoData
    else:
        aggregateColoradoLabelData_AvI = pandas.concat([aggregateColoradoLabelData_AvI, coloradoData])
    i += 1
        
#aggregateColoradoLabelData_AvI = aggregateColoradoLabelData_AvI.set_index("Tweet_ID")



i=0
aggregateColoradoLabelData_RvN = None
for filename in tweetLabelFiles_RvN:
    coloradoData = pandas.read_csv(filename, sep="\t", header=None)
    coloradoData.columns = ["Tweet_ID","Related_Label"]
    #coloradoData.set_index("Tweet_ID")    ////This doesnt work for some reason...
    if i == 0:
        aggregateColoradoLabelData_RvN = coloradoData
    else:
        aggregateColoradoLabelData_RvN = pandas.concat([aggregateColoradoLabelData_RvN, coloradoData])
    i += 1
        
#aggregateColoradoLabelData_RvN = aggregateColoradoLabelData_RvN.set_index("Tweet_ID")


i=0
aggregateColoradoLabelData_SvO = None
for filename in tweetLabelFiles_SvO:
    coloradoData = pandas.read_csv(filename, sep="\t", header=None)
    coloradoData.columns = ["Tweet_ID","Self_Label"]
    #coloradoData.set_index("Tweet_ID")    ////This doesnt work for some reason...
    if i == 0:
        aggregateColoradoLabelData_SvO = coloradoData
    else:
        aggregateColoradoLabelData_SvO = pandas.concat([aggregateColoradoLabelData_SvO, coloradoData])
    i += 1
        
#aggregateColoradoLabelData_SvO = aggregateColoradoLabelData_SvO.set_index("Tweet_ID")
```
